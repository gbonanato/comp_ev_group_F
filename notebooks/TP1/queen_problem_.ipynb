```python
from TP.core.logging.observer import BestFitnessLogger
from TP.core.logging.progress import EAProgressLogger
from TP.problems.queens.interface import QueenProblemOrchestrator

generations = []
for _ in range(1000):
    data_logger = BestFitnessLogger()
    process_logger = EAProgressLogger()

    orchestrator = QueenProblemOrchestrator(
        board_size=8,
        pop_size=20,
        observers=[data_logger],
        loggers=[process_logger],
    )
    out = orchestrator.run()
    generations.append(data_logger.generations[-1])
```

```python
out
```

```python
mean = sum(generations) / len(generations)
mean
```

```python
import statistics

min_val = min(generations)
max_val = max(generations)
mean_val = statistics.mean(generations)
median_val = statistics.median(generations)
std_dev = statistics.stdev(generations)

print(f'Std Dev: {round(std_dev, 2)}')
print(f'Min: {min_val}')
print(f'Max: {max_val}')
print(f'Mean: {mean_val}')
print(f'Median: {median_val}')
```

```python
import matplotlib.pyplot as plt

plt.plot(data_logger.generations, data_logger.best_fitness)
plt.xlabel('Geração')
plt.ylabel('Fitness do melhor indivíduo')
plt.title('Convergência do Algoritmo Genético')
plt.grid(True)
plt.show()
```
