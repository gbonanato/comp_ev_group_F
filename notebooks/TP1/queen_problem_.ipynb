```python
from TP.core.logging.observer import BestFitnessLogger
from TP.core.logging.progress import EAProgressLogger
from TP.problems.queens.interface import QueenProblemOrchestrator

generations = []
for _ in range(1000):
    data_logger = BestFitnessLogger()
    process_logger = EAProgressLogger()

    orchestrator = QueenProblemOrchestrator(
        board_size=8,
        pop_size=20,
        observers=[data_logger],
        loggers=[process_logger],
    )
    out = orchestrator.run()
    generations.append(data_logger.generations[-1])
```

```python
mean = sum(generations) / len(generations)
mean
```
